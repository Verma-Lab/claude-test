# PMBB Specimen Counts
Counts of specimens by type from `biobank_analytics.dev_phi.pmbb_specimens`.

```python
df = spark.table("biobank_analytics.dev_phi.pmbb_specimens")
```

```python
from pyspark.sql.functions import col

specimen_counts = (
    df.groupBy("specimen_type")
    .count()
    .withColumnRenamed("count", "specimen_count")
    .orderBy(col("specimen_count").desc())
)

specimen_counts.display()
```

```python
total = df.count()
print(f"Total specimens: {total:,}")
```
